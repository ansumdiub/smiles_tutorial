```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import PyMol
# RDKit is a popular open source toolkit for cheminformatics and machine learning
```

```python
# https://commons.healthymaterials.net/chemicals/2005314
# 2,4-IMIDAZOLIDINEDIONE, 1,3-BIS(HYDROXYMETHYL)-5,5-DIMETHYL-
# Benchmark 1 "Avoid - Chemical of High Concern"
m = Chem.MolFromSmiles("CC1(C(=O)N(C(=O)N1CO)CO)C")
m
```

```python
# Canonical SMILES
# There is no universal canonical SMILES. Every toolkit uses a different algorithm.
s = Chem.MolToSmiles(m)
print(s)
```

```python
# To get good 3D conformations, itâ€™s almost always a good idea to add hydrogens to the molecule first
m1 = Chem.AddHs(m)

# embedding a molecule, ETKDG is the default conformation generation method.
AllChem.EmbedMolecule(m1) 
```

```python
# a very convenient method is to connect RDKit to a running instance of PyMOL, 
# using a communication protocol called xmlrpc
# You need to start PyMOL using a special option to start the xmlrpc server,
# copy and paste below in terminal
# /Applications/MacPyMOL.app/Contents/MacOS/MacPyMOL -R

v = PyMol.MolViewer()
v.ShowMol(m1, name='embedded', showOnly=True)
v.server.do('ray')
v.GetPNG()
```
